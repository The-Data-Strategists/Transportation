## Exploratory Data Analysis (EDA)...

```python
# Calculate total fare (EGP) for each distance_km to analyze the relation between distance and revenue
subs_by_distance = All.groupby('distance_km')['fare_EGP'].sum()
print(subs_by_distance)
```

```python
# Assign fuel type to each car based on model and manufacturing year..
def assign_fuel_type(row):
    if row['car_model'] in ['Chevrolet', 'Nissan']:
        return '92'
    elif row['car_model'] in ['Toyota', 'Hyundai', 'Kia']:
        return '92' if row['car_year'] <= 2015 else '95'
    else:
        return '92'

All['fuel_type'] = All.apply(assign_fuel_type, axis=1)
```

```python
# Calculate trip fare (EGP) based on fuel type, fuel price, and trip distance
All['fare_egp'] = All.apply(
    lambda row: row['distance_km'] * row['octane92_price'] if row['fuel_type'] == '92'
    else row['distance_km'] * row['octane95_price'],
    axis=1
)
```

```python
  # Categorize customers into age groups based on their age
def category_age(age):
    if   18 <= age <= 24 :
        return 'Young_Adult'
    elif  25 <= age <= 39:
        return 'Adult'
    elif  40 <= age <= 59:
        return 'Middel_age-Adult'
    elif  60 <= age <= 64:
        return 'Older_Adult'
    else:
        return 'Out of range'

All['age_category'] = All['age'].apply(category_age)
```

```python
# Show the number of customers in each age category to understand age distribution
All['age_category'].value_counts()
```

```python
All.groupby('city_area')['age_category'].value_counts()
```

```python
 # Categorize ratings into levels (Excellent, Very Good, Good, etc.) based on rating value
def category_rating(rating): 
    if 4.5 <= rating <= 5:
        return 'Excellent' 
    elif 3.9 <= rating <= 4.45:
        return 'Very Good'
    elif 3.5 <= rating <= 3.88: 
        return 'Good'  
    else:
        return 'Out of range'

All['rating_category'] = All['rating'].apply(category_rating)
```

```python
# Show the distribution of rating categories for each car model to analyze performance by model
All.groupby('car_model')['rating_category'].value_counts()
```

```python
# Calculate the correlation between distance and fare to see how strongly they are related
All[["distance_km","fare_EGP"]].corr()
```

## Data Visualization & Dashboard.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
# Scatter plot to visualize the relationship between distance (km) and fare (EGP)
warnings.filterwarnings('ignore')
# Scatter plot showing relationship between distance and fare
plt.figure(figsize=(8,5))
sns.scatterplot(data=All, x='distance_km', y='fare_EGP', color='royalblue')

plt.title('Relationship between Distance (km) and Fare (EGP)', fontsize=14)
plt.xlabel('Distance (km)', fontsize=12)
plt.ylabel('Fare (EGP)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Bar chart showing the number of records per city area by year, with values displayed above the bars
warnings.filterwarnings('ignore')
data_counts = All.groupby(['city_area', 'year']).size().reset_index(name='count')
plt.figure(figsize=(10,6))
ax = sns.barplot(data=data_counts, x='city_area', y='count', hue='year', palette='Reds')
plt.title('Number of Records per City by Year', fontsize=14)
plt.xlabel('City Area', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Year', loc='upper center', ncol=2)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10, padding=3)
plt.tight_layout()
plt.show()
```

```python
# Calculate the count of each age category
counts = All['age_category'].value_counts()
# Plot bar chart for age category distribution
plt.figure(figsize=(8, 5))
bars = counts.plot(kind='bar', color=['#ff9999', '#ff9999', '#ff9999', '#ff9999'])
plt.title('Distribution of Age Categories', fontsize=14)
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels above bars
for i, v in enumerate(counts):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.show()
```

```python
# Pie chart showing the percentage and total fare distribution by gender using Plotly
import plotly.express as px
data = pd.DataFrame({
    'gender': ['Female', 'Male'],
    'fare_egp': [102242.6975, 107256.4500]
})
fig = px.pie(
    data,
    names='gender',
    values='fare_egp',
    title='Total Fare (EGP) by Gender',
    color='gender',
    color_discrete_map={'Female': '#ff9999', 'Male': '#66b3ff'},
    hole=0.2 
)
fig.update_traces(textinfo='percent+label', pull=[0.05, 0]) 
fig.show()
```
