````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ÿ≠ÿ≥ÿßÿ® ÿπÿØÿØ ÿßŸÑŸÖÿ≠ÿ∑ÿßÿ™ Ÿàÿ•ÿ¨ŸÖÿßŸÑŸä ÿßŸÑÿ±ŸÉÿßÿ®
total_stations = df_Metro_Ridership['station'].nunique()
total_passengers = df_Metro_Ridership['passengers'].sum()

# ÿ•ŸÜÿ¥ÿßÿ° ŸÖÿÆÿ∑ÿ∑ Ÿäÿ≠ÿ™ŸàŸä ÿ®ÿ∑ÿßŸÇÿ™ŸäŸÜ
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type": "indicator"}, {"type": "indicator"}]]
)

# ÿ®ÿ∑ÿßŸÇÿ© ÿπÿØÿØ ÿßŸÑŸÖÿ≠ÿ∑ÿßÿ™
fig.add_trace(go.Indicator(
    mode="number",
    value=total_stations,
    title={"text": "üìç ÿπÿØÿØ ÿßŸÑŸÖÿ≠ÿ∑ÿßÿ™"},
    number={'valueformat': ',.0f'},
), row=1, col=1)

# ÿ®ÿ∑ÿßŸÇÿ© ÿ•ÿ¨ŸÖÿßŸÑŸä ÿßŸÑÿ±ŸÉÿßÿ®
fig.add_trace(go.Indicator(
    mode="number",
    value=total_passengers,
    title={"text": "üë• ÿ•ÿ¨ŸÖÿßŸÑŸä ÿπÿØÿØ ÿßŸÑÿ±ŸÉÿßÿ®"},
    number={'valueformat': ',.0f'},
), row=1, col=2)

# ÿßŸÑÿ™ŸÜÿ≥ŸäŸÇ ÿßŸÑÿπÿßŸÖ
fig.update_layout(
    height=250,
    width=600,
    margin=dict(l=20, r=20, t=40, b=20),
    paper_bgcolor="#f7f7f7",
    font=dict(size=16, color="black"),
)

fig.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")

monthly_passengers = df_Metro_Ridership.groupby('month')['passengers'].sum().reset_index()

max_month = monthly_passengers.loc[monthly_passengers['passengers'].idxmax()]

plt.figure(figsize=(10,6))

sns.lineplot(
    data=monthly_passengers,
    x='month',
    y='passengers',
    marker='o',
    linewidth=2.5,
    markersize=8,
    color='#005BBB'
)

plt.scatter(
    max_month['month'],
    max_month['passengers'],
    color='crimson',
    s=120,
    zorder=5,
    label=f"Peak Month ({max_month['month']})"
)

# ‚≠ê ÿ™ŸÉÿ®Ÿäÿ± ÿßŸÑÿ£ÿ±ŸÇÿßŸÖ ŸÅŸàŸÇ ÿßŸÑŸÜŸÇÿßÿ∑ ŸáŸÜÿß (fontsize=12)
for i, row in monthly_passengers.iterrows():
    plt.text(row['month'], row['passengers']+1000,
             f"{row['passengers']:,}",
             ha='center', fontsize=12, weight='bold')

plt.title("üìà Monthly Passengers Trend", fontsize=16, weight='bold')
plt.xlabel("Month", fontsize=13)
plt.ylabel("Total Passengers", fontsize=13)

plt.xticks(np.arange(1, 13))
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()

sns.despine()

plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ÿ™ÿ¨ŸÖŸäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ≠ÿ≥ÿ® ÿßŸÑŸÖÿ≠ÿ∑ÿ©
station_passengers = df_Metro_Ridership.groupby('station')['passengers'].sum().reset_index()

# ÿ™ÿ±ÿ™Ÿäÿ® ÿßŸÑŸÖÿ≠ÿ∑ÿßÿ™ ŸÖŸÜ ÿßŸÑÿ£ŸÉÿ®ÿ± ŸÑŸÑÿ£ÿµÿ∫ÿ±
station_passengers = station_passengers.sort_values(by='passengers', ascending=False)

# ÿ™ÿµŸÖŸäŸÖ ÿßÿ≠ÿ™ÿ±ÿßŸÅŸä
plt.figure(figsize=(12,6))
sns.set_theme(style="whitegrid")

# ÿ±ÿ≥ŸÖ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿ®ÿ£ŸÑŸàÿßŸÜ ŸÖÿ™ŸÜÿßÿ≥ŸÇÿ©
colors = sns.color_palette("Blues_r", len(station_passengers))
ax = sns.barplot(
    data=station_passengers,
    x='station',
    y='passengers',
    palette=colors
)

# ‚≠ê ÿ•ÿ∂ÿßŸÅÿ© ÿ£ÿ±ŸÇÿßŸÖ ÿßÿ≠ÿ™ÿ±ÿßŸÅŸäÿ© ŸÅŸàŸÇ ÿßŸÑÿ£ÿπŸÖÿØÿ© ŸÖÿπ Box ÿÆŸÑŸÅŸáÿß
for i, row in station_passengers.iterrows():
    ax.text(
        i,
        row['passengers'] + max(station_passengers['passengers'])*0.01,
        f"{row['passengers']:,}",
        ha='center',
        fontsize=12,
        fontweight='bold',
        color='black',
        bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.2', alpha=0.9)
    )

# ÿ™ŸÜÿ≥ŸäŸÇÿßÿ™ ŸÜŸáÿßÿ¶Ÿäÿ©
plt.title("üöá Total Passengers per Metro Station", fontsize=18, weight='bold')
plt.xlabel("Station", fontsize=13, weight='bold')
plt.ylabel("Total Passengers", fontsize=13, weight='bold')

plt.xticks(rotation=30, fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.4)
sns.despine()

plt.tight_layout()
plt.show()
````
