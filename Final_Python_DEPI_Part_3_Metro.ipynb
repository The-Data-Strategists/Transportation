```python
import matplotlib.pyplot as plt
import seaborn as sns

# حساب إجمالي عدد الركاب في كل شهر
monthly_passengers = df_Metro_Ridership.groupby(['month'])['passengers'].sum().reset_index()

# رسم الخط البياني
plt.figure(figsize=(8,5))
sns.lineplot(data=monthly_passengers, x='month', y='passengers', marker='o', color='royalblue')

plt.title('Monthly Passengers Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Passengers', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
