```python
def category_age(age):
    if   18 <= age <= 24 :
        return 'Young_Adult'
    elif  25 <= age <= 39:
        return 'Adult'
    elif  40 <= age <= 59:
        return 'Middel_age-Adult'
    elif  60 <= age <= 64:
        return 'Older_Adult'
    else:
        return 'Out of range'

Smart['age_category'] = Smart['age'].apply(category_age)
```

```python
def category_rating(rating): 
    if 4.5 <= rating <= 5:
        return 'Excellent' 
    elif 3.9 <= rating <= 4.45:
        return 'Very Good'
    elif 3.5 <= rating <= 3.88: 
        return 'Good'  
    else:
        return 'Out of range'

Smart['rating_category'] = Smart['rating'].apply(category_rating)
```

```python
def assign_fuel_type(row):
    if row['car_model'] in ['Chevrolet', 'Nissan']:
        return '92'
    elif row['car_model'] in ['Toyota', 'Hyundai', 'Kia']:
        return '92' if row['car_year'] <= 2015 else '95'
    else:
        return '92'

Smart['fuel_type'] = Smart.apply(assign_fuel_type, axis=1)
```

```python
Smart['fare_egp'] = Smart.apply(
    lambda row: row['distance_km'] * row['octane92_price'] if row['fuel_type'] == '92'
    else row['distance_km'] * row['octane95_price'],
    axis=1
)
```
